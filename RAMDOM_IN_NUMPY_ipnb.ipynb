```python
from numpy import random

x = random.choice([3,4,5,6], size=(3,5))

print(x)
```

```python
from numpy import random

x = random.choice([3,4,5,9], p=[0.1,0.3,0.6,0.0])

print(x)
```

```python
from numpy import random

x = random.choice([3,4,5,9], p=[0.1,0.3,0.6,0.0], size=(3,5))

print(x)
```

```python
from numpy import random
import numpy as np

arr = np.array([1,2,3,4,5])

random.shuffle(arr)

print(arr)
```

```python
from numpy import random
import numpy as np

arr = np.array([1,2,3,4,5])

print(random.permutation(arr))
```
